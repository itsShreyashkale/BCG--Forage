```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

client_df = pd.read_csv("client_data (1).csv")
price_df = pd.read_csv("price_data (1).csv")
```

```python
client_df.head()
price_df.head()
```

```python
client_df.info()
price_df.info()
```

```python
date_cols = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']

for col in date_cols:
    client_df[col] = pd.to_datetime(client_df[col])
```

```python
price_df['price_date'] = pd.to_datetime(price_df['price_date'])
```

```python
client_df.describe()
price_df.describe()
```

```python
num_cols = client_df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure()
    sns.histplot(client_df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
sns.countplot(x='churn', data=client_df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='churn', data=client_df)
plt.title("Churn Distribution")
plt.show()
```

```python
cat_cols = client_df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure()
    client_df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f'Top categories in {col}')
    plt.show()
```
